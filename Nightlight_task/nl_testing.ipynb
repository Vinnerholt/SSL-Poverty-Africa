```python
import torch
import numpy as np
import random
```

```python
preds = torch.Tensor([0,0,0,0])
labels = torch.Tensor([1, 1, 1, 1])

criterion = torch.nn.BCEWithLogitsLoss()
loss = criterion(preds, labels)
print(loss)
```

```python
sigmoid = torch.nn.Sigmoid()
preds = torch.Tensor([0,0,0,0])
labels = torch.Tensor([1, 1, 1, 1])
preds = sigmoid(preds)
print(preds)

criterion = torch.nn.BCELoss()
loss = criterion(preds, labels)
print(loss)
```

```python
from nightlight_ssl_dataset import NightlightSSLDataset
from rasterio.plot import show
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## NL Images are normalized before plotted (adjust=True in show()), therefore some are completely white or black
## Swapped: 0 means no swap, and Swapped: 1 means NL has been swapped

train_csv = "/cephyr/NOBACKUP/groups/globalpoverty1/JesperBenjamin/imagepath1000.csv"

dataset = NightlightSSLDataset(train_csv)
loader= torch.utils.data.DataLoader(dataset, batch_size=2, shuffle=True)

for i, (images, labels) in enumerate(loader):
    print(images.size())
    img = images[0].detach().numpy()
    show(img[(2,1,0),:,:], adjust=True)
    show(img[(7,7,7),:,:], adjust=True)
    print("Swapped: ", labels[0].item())
    
    if i >= 10:
        break
```
